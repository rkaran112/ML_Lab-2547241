## Experiment 1: CIA Percentage vs GPA
This experiment aims to see if a student's CIA percentage can predict their GPA.

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# PART A - DATA CLEANING
# 1. Load CSV using pandas
file_path = '/content/student_survey.qcsv'
df = pd.read_csv(file_path)

# 2. Display first 5 rows
print("First 5 rows:")
display(df.head())

# 3. Check shape
print(f"Original shape: {df.shape}")

# 4. Find missing values
print("Missing values:")
print(df.isnull().sum())

# 5. Remove missing values
df.dropna(inplace=True)

# 6. Convert columns to numeric and rename
# We rename the specific column as requested
df.rename(columns={'Your CIA % of last semester': 'CIA_percentage',
                   'Your maximum attendance % till last semester': 'Attendance_percentage',
                   'Your GPA of last semester': 'GPA'}, inplace=True)

df['CIA_percentage'] = pd.to_numeric(df['CIA_percentage'], errors='coerce')
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
df.dropna(subset=['CIA_percentage', 'GPA'], inplace=True)

# 7. Remove duplicates
df.drop_duplicates(inplace=True)

# 8. Summary statistics
print("Summary Statistics:")
display(df[['CIA_percentage', 'GPA']].describe())
```

```python
# PART B - SCIKIT-LEARN METHOD
# 1. Split data
X = df[['CIA_percentage']]
y = df['GPA']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Create model
model = LinearRegression()

# 3. Train model
model.fit(X_train, y_train)

# 4. Get slope and intercept
skl_slope = model.coef_[0]
skl_intercept = model.intercept_

# 5. Make predictions
y_pred_sklearn = model.predict(X_test)

# 6. Print results
print(f"Sklearn Slope: {skl_slope:.4f}")
print(f"Sklearn Intercept: {skl_intercept:.4f}")
print("First 5 Sklearn predictions:", y_pred_sklearn[:5])
```

```python
# PART C - MANUAL OLS METHOD
x_manual = X_train['CIA_percentage']
y_manual = y_train

# Formulas
mean_x = np.mean(x_manual)
mean_y = np.mean(y_manual)

num = np.sum((x_manual - mean_x) * (y_manual - mean_y))
den = np.sum((x_manual - mean_x)**2)

manual_slope = num / den
manual_intercept = mean_y - (manual_slope * mean_x)

# Predictions on test set
y_pred_manual = (manual_slope * X_test['CIA_percentage']) + manual_intercept

print(f"Manual Slope: {manual_slope:.4f}")
print(f"Manual Intercept: {manual_intercept:.4f}")
print("First 5 Manual predictions:", y_pred_manual.values[:5])
```

```python
# PART D - COMPARISON
comparison = pd.DataFrame({
    'Actual': y_test.head(10),
    'Sklearn': y_pred_sklearn[:10],
    'Manual': y_pred_manual.head(10).values
})
comparison['Difference'] = abs(comparison['Sklearn'] - comparison['Manual'])
display(comparison)

avg_diff = comparison['Difference'].mean()
print(f"Average Difference: {avg_diff}")
print("Are both methods giving the same results? YES" if avg_diff < 1e-6 else "NO")
```

```python
# PART E - SAVE TO PICKLE
weights = {'slope': manual_slope, 'intercept': manual_intercept}
with open('linear_regression_weights.pkl', 'wb') as f:
    pickle.dump(weights, f)

# Load and verify
with open('linear_regression_weights.pkl', 'rb') as f:
    loaded = pickle.load(f)

# Prediction with loaded values
test_val = 75
pred = (loaded['slope'] * test_val) + loaded['intercept']
print(f"Prediction for 75% CIA using loaded weights: {pred:.2f}")
```

## Experiment 2: Attendance Percentage vs GPA
Now we repeat the process using Attendance as the input variable.

```python
# Experiment 2: Attendance -> GPA
# Step 1: Clean and convert Attendance data
# We remove '%' signs and ensure the column is numeric before modeling
df['Attendance_percentage'] = df['Attendance_percentage'].astype(str).str.replace('%', '')
df['Attendance_percentage'] = pd.to_numeric(df['Attendance_percentage'], errors='coerce')
df.dropna(subset=['Attendance_percentage'], inplace=True)

X2 = df[['Attendance_percentage']]
y2 = df['GPA']

# Split data
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=42)

# Sklearn method
model2 = LinearRegression().fit(X2_train, y2_train)
skl_slope2 = model2.coef_[0]
skl_intercept2 = model2.intercept_

# Manual OLS method
x2_m = X2_train['Attendance_percentage']
y2_m = y2_train
m_slope2 = np.sum((x2_m - x2_m.mean())*(y2_m - y2_m.mean())) / np.sum((x2_m - x2_m.mean())**2)
m_intercept2 = y2_m.mean() - (m_slope2 * x2_m.mean())

print(f"Exp 2 Sklearn Slope: {skl_slope2:.4f}")
print(f"Exp 2 Manual Slope: {m_slope2:.4f}")
```

```python
# PART C - MANUAL OLS METHOD (Experiment 2)
# Prediction = slope * x + intercept using the manual weights calculated previously
y_pred_manual2 = (m_slope2 * X2_test['Attendance_percentage']) + m_intercept2

# Also get scikit-learn predictions for comparison
y_pred_sklearn2 = model2.predict(X2_test)

print(f"Manual OLS Slope (Exp 2): {m_slope2:.4f}")
print(f"Manual OLS Intercept (Exp 2): {m_intercept2:.4f}")
print("First 5 Manual OLS predictions (Exp 2):", y_pred_manual2.values[:5])
```

```python
# PART D - COMPARISON (Experiment 2)
comparison2 = pd.DataFrame({
    'Actual GPA': y2_test.head(10),
    'Sklearn Predicted': y_pred_sklearn2[:10],
    'Manual Predicted': y_pred_manual2.head(10).values
})
comparison2['Difference'] = abs(comparison2['Sklearn Predicted'] - comparison2['Manual Predicted'])

print("Comparison of Sklearn vs Manual predictions for Experiment 2:")
display(comparison2)

avg_diff2 = comparison2['Difference'].mean()
print(f"Average absolute difference: {avg_diff2:.6f}")
print("Are both methods giving the same results?", "YES" if avg_diff2 < 1e-6 else "NO")
```

```python
# PART E - SAVE TO PICKLE (Experiment 2)
# Save the learned weights to a new pickle file
weights2 = {
    'slope': m_slope2,
    'intercept': m_intercept2
}

pickle_file_exp2 = 'linear_regression_weights_exp2.pkl'
with open(pickle_file_exp2, 'wb') as f:
    pickle.dump(weights2, f)

print(f"Experiment 2 weights saved to {pickle_file_exp2}")

# Final Summary Output
print("\n--- Experiment 2 Summary ---")
print("Experiment Name: Attendance percentage predicting GPA")
print("Input Variable (X): Attendance Percentage")
print("Output Variable (Y): GPA")
print(f"Sklearn Slope: {skl_slope2:.4f}, Intercept: {skl_intercept2:.4f}")
print(f"Manual Slope: {m_slope2:.4f}, Intercept: {m_intercept2:.4f}")
```
