## Missing Value Treatment Strategy

Based on the initial inspection, both datasets contain missing values that need to be addressed before further analysis. For each column with missing data, I will do the following: dropping the column, dropping the rows, or imputing the missing values. The choice will be justified based on the column's nature, the percentage of missing values, and its potential importance for the model.

### `city_day.csv` - Missing Value Treatment

First, I will convert the 'Date' column to a datetime object, as it is currently an object type. This is crucial for time-series analysis and any potential future feature engineering based on time.

```python
# Convert 'Date' column to datetime objects
city_day_df['Date'] = pd.to_datetime(city_day_df['Date'])
```

#### Missing Values Before Treatment for `city_day_df`

```python
print("Missing values in city_day_df before treatment:")
display(city_day_df.isnull().sum().sort_values(ascending=False))
```

#### Strategy and Justification for `city_day_df` Columns:

1.  **`Xylene`**: This column has a very high percentage of missing values (~60%). Imputing such a large portion of the data could introduce significant bias and distort the underlying distribution. Therefore, it's safer to **drop this column**.
2.  **`PM10`, `NH3`, `Toluene`, `NOx`, `AQI`, `O3`, `SO2`, `NO`, `NO2`, `CO`, `PM2.5`, `Benzene`**: These are all numerical features representing air pollutant concentrations or the Air Quality Index. They have varying but significant percentages of missing values (ranging from ~10% to ~38%). Given their importance in understanding air quality, dropping rows would lead to substantial data loss. **Median imputation** is chosen over mean imputation because air quality data can often be skewed (e.g., occasional high pollution events), and the median is more robust to outliers and preserves the distribution shape better than the mean.
3.  **`AQI_Bucket`**: This is a categorical column directly related to `AQI`. Since `AQI` will be imputed, `AQI_Bucket` might still have missing values. For categorical features, **mode imputation** (filling with the most frequent category) is a common and appropriate strategy as it does not assume any distribution and maintains the integrity of the categorical variable.

```python
# 1. Drop 'Xylene' column
city_day_df = city_day_df.drop(columns=['Xylene'])

# 2. Impute numerical columns with their median
numerical_cols_to_impute = ['PM10', 'NH3', 'Toluene', 'NOx', 'AQI', 'O3', 'SO2', 'NO', 'NO2', 'CO', 'PM2.5', 'Benzene']
for col in numerical_cols_to_impute:
    if col in city_day_df.columns:
        city_day_df[col] = city_day_df[col].fillna(city_day_df[col].median())

# 3. Impute 'AQI_Bucket' with its mode
if 'AQI_Bucket' in city_day_df.columns:
    city_day_df['AQI_Bucket'] = city_day_df['AQI_Bucket'].fillna(city_day_df['AQI_Bucket'].mode()[0])
```

#### Missing Values After Treatment for `city_day_df`

```python
print("Missing values in city_day_df after treatment:")
display(city_day_df.isnull().sum().sort_values(ascending=False))
```

# Task 5: Handling Extreme AQI Values

## Identifying and Treating Extreme AQI Values

This task focuses on addressing the presence of implausibly high AQI readings in the dataset. Such extreme values can distort statistical analyses and machine learning models. I will:

1.  **Identify Extreme Values**: Use visual methods (box plot) and statistical methods (IQR) to detect outliers in the 'AQI' column.
2.  **Quantify Affected Values**: Count how many records are considered extreme.
3.  **Choose and Justify Treatment**: Select an appropriate method (e.g., removal, winsorization) and explain the rationale.
4.  **Apply Treatment**: Implement the chosen method.
5.  **Demonstrate Effectiveness**: Provide a visual comparison and summary statistics before and after treatment.

### Step 1: Visual Identification of Extremes (Box Plot)

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(y=city_day_df['AQI'], color='lightcoral')
plt.title('Box Plot of AQI Values (Before Treatment)')
plt.ylabel('AQI Value')
plt.show()

print("Descriptive statistics of AQI before treatment:")
display(city_day_df['AQI'].describe())
```

### Step 2: Quantify Extremes using the IQR Method

The box plot clearly shows a significant number of outliers above the main distribution. To formally quantify these 'extreme' values, I will use the Interquartile Range (IQR) method. Values falling above `Q3 + 1.5 * IQR` will be considered outliers. This method is robust to skewed distributions and is commonly used for outlier detection.

```python
Q1 = city_day_df['AQI'].quantile(0.25)
Q3 = city_day_df['AQI'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR

extreme_aqi_values = city_day_df[city_day_df['AQI'] > upper_bound]
num_extreme_values = len(extreme_aqi_values)

print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"Upper bound for outliers (Q3 + 1.5 * IQR): {upper_bound:.2f}")
print(f"Number of extreme AQI values identified: {num_extreme_values}")
print(f"Percentage of extreme values: {(num_extreme_values / len(city_day_df)) * 100:.2f}%")

print("\nExamples of extreme AQI values:")
display(extreme_aqi_values[['City', 'Date', 'AQI']].head())
```

### Step 3: Justification for Treatment Method

Given the data quality report's indication of 'implausibly high' values that would 'distort every statistic and model', simply deleting these records might lead to significant data loss (as seen, {num_extreme_values} records, which is `{(num_extreme_values / len(city_day_df)) * 100:.2f}%` of the dataset). This could remove valuable information from the non-AQI columns for these days.

Instead, **winsorization (or capping)** is a more appropriate method. It involves replacing outliers with a specified percentile value (or a predefined plausible maximum). This approach:

*   **Retains Data Points**: No records are lost, preserving other potentially useful features for those dates and cities.
*   **Mitigates Distortion**: The impact of extreme values on the mean, standard deviation, and model training is significantly reduced.
*   **Reflects Reality (to an extent)**: It assumes that while the recorded value is implausible, there was still a high pollution event, which is then capped at a more believable, yet high, level.

I will cap the identified extreme AQI values at the calculated `upper_bound` (`{upper_bound:.2f}`). This ensures that any AQI above this threshold is replaced by `{upper_bound:.2f}`, making them high but no longer 'implausibly' so, based on the distribution's characteristics.

```python
# Apply winsorization (capping) to the 'AQI' column
city_day_df_cleaned = city_day_df.copy() # Create a copy to show before/after comparison
city_day_df_cleaned['AQI'] = city_day_df_cleaned['AQI'].apply(lambda x: min(x, upper_bound))

print("AQI values successfully capped at the upper bound.")
```

### Step 4: Visual Comparison and Quantification After Treatment

To demonstrate that the data is now cleaner, I will display the descriptive statistics and a box plot of the 'AQI' column after winsorization, comparing them with the 'before' state.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=city_day_df_cleaned['AQI'], color='lightgreen')
plt.title('Box Plot of AQI Values (After Treatment - Winsorized)')
plt.ylabel('AQI Value')
plt.show()

print("Descriptive statistics of AQI after treatment:")
display(city_day_df_cleaned['AQI'].describe())

print("\nComparison of Max AQI values before and after treatment:")
print(f"Max AQI before treatment: {city_day_df['AQI'].max():.2f}")
print(f"Max AQI after treatment: {city_day_df_cleaned['AQI'].max():.2f}")

print(f"Count of values affected by capping: {num_extreme_values}")
```

The box plot after treatment shows the absence of outliers, and the maximum value in the descriptive statistics has been reduced significantly to the defined upper bound. This demonstrates that the winsorization successfully addressed the implausibly high AQI readings, resulting in a cleaner dataset for further analysis.

## Analyzing Air Quality Trends Over Time

To understand whether air quality has improved, worsened, or stayed the same, especially in relation to policies introduced after 2018, I will first extract the year from the 'Date' column. Then, I will calculate the average Air Quality Index (AQI) for each year and visualize this trend.

```python
# Extract the year from the 'Date' column
city_day_df_cleaned['Year'] = city_day_df_cleaned['Date'].dt.year

# Calculate the average AQI per year
aqi_yearly_avg = city_day_df_cleaned.groupby('Year')['AQI'].mean().reset_index()

print("Average AQI per Year:")
display(aqi_yearly_avg.head())
```

### Visualizing Yearly AQI Trends

I will use a line plot to visually represent the average AQI over the years. This type of visualization is excellent for showing trends and making them immediately readable to a non-data scientist audience.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

sns.lineplot(x='Year', y='AQI', data=aqi_yearly_avg, marker='o', color='indianred', linewidth=2)

plt.title('Average Air Quality Index (AQI) Trend Over Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average AQI', fontsize=12)
plt.xticks(aqi_yearly_avg['Year'], rotation=45) # Ensure all years are shown and readable
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Identify the most and least polluted years
most_polluted_year = aqi_yearly_avg.loc[aqi_yearly_avg['AQI'].idxmax()]
least_polluted_year = aqi_yearly_avg.loc[aqi_yearly_avg['AQI'].idxmin()]

print(f"The most polluted year was {most_polluted_year['Year']} with an average AQI of {most_polluted_year['AQI']:.2f}.")
print(f"The least polluted year was {least_polluted_year['Year']} with an average AQI of {least_polluted_year['AQI']:.2f}.")
```

## Task 8: Can the two datasets talk to each other?

### Revised Data Transformation and Merging Strategy

The initial attempt to merge the `city_day_df_cleaned` (daily city AQI) and `crop_production_df` (annual state crop production) by aggregating to a common **State-Year** level resulted in an empty combined dataset. This was due to a significant lack of overlapping (State, Year) combinations between the two datasets, primarily a timeframe mismatch and data sparsity across states for the single overlapping year (2015).

To overcome this limitation and enable a meaningful analysis of whether states with worse air quality also tend to produce less crop, I will revise the aggregation strategy:

**Revised Transformation Needed:**

1.  **Aggregate `city_day_df_cleaned` to State-Level (Overall Average AQI)**: Instead of yearly AQI, I will calculate the *overall average* AQI for each `State_Name` across all available years. This will provide a general measure of air quality for each state.

2.  **Aggregate `crop_production_df` to State-Level (Overall Total Production and Area)**: Similarly, I will calculate the *overall total* `Production` and `Area` for each `State_Name` across all available `Crop_Year`s. This will provide a summary of agricultural output for each state.

**Why this revised transformation is necessary:**

By aggregating to a state-only level (removing the year as a merge key), we maximize the chances of finding common states between the two datasets, thereby ensuring a non-empty combined dataset suitable for correlation analysis. This approach allows us to answer the question about the general relationship between a state's typical air quality and its agricultural capacity, even if we lose the year-over-year dynamic.

**Merge Operation:**

Once both datasets are aggregated to the state level, they will be merged using an `inner` join on the `State_Name` column. This will create a combined dataset where each row represents a state, with its average AQI and total crop production/area data.

```python
# 1. Aggregate city_day_df_cleaned to State-Level (overall average AQI)
state_avg_aqi_df = city_day_df_cleaned.groupby('State_Name')['AQI'].mean().reset_index()
print("Overall Average AQI per State (first 5 rows):")
display(state_avg_aqi_df.head())

# 2. Aggregate crop_production_df to State-Level (overall total Production and Area)
state_total_crop_df = crop_production_df.groupby('State_Name').agg(
    Overall_Total_Production=('Production', 'sum'),
    Overall_Total_Area=('Area', 'sum')
).reset_index()
print("\nOverall Total Crop Production and Area per State (first 5 rows):")
display(state_total_crop_df.head())

# 3. Merge the two aggregated dataframes on 'State_Name'
combined_state_df = pd.merge(
    state_avg_aqi_df,
    state_total_crop_df,
    on='State_Name',
    how='inner'
)

print("\nCombined State-Level DataFrame (first 5 rows):")
display(combined_state_df.head())
print(f"Combined State-Level DataFrame shape: {combined_state_df.shape}")
```

### Visualizing Relationships Across Numerical Features (State-Level)

With the datasets now merged at the state level, I will generate a correlation matrix and visualize it as a heatmap to explore the relationships between the numerical variables: `AQI`, `Overall_Total_Production`, and `Overall_Total_Area`. This will help us understand general correlations across states.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numerical columns for correlation analysis from the combined_state_df
numerical_cols = ['AQI', 'Overall_Total_Production', 'Overall_Total_Area']
correlation_matrix = combined_state_df[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=.5
)
plt.title('Correlation Matrix of State-Level Variables (Average AQI, Total Production, Total Area)', fontsize=12)
plt.show()
```

### Two Most Interesting Relationships and Their Real-World Reasons (State-Level)

Based on the correlation matrix derived from the state-level aggregated data, here are two interesting relationships:

1.  **Relationship between `AQI` and `Overall_Total_Production` (Correlation: -0.25)**:
    *   **What it suggests**: There is a weak negative correlation between a state's overall average AQI and its overall total crop production. This indicates that states with, on average, worse air quality (higher AQI) tend to have slightly lower total crop production over the entire period, and vice-versa.
    *   **Why it might exist**: Poor air quality, characterized by higher AQI, can negatively impact crop yields. Air pollutants such as ground-level ozone, particulate matter, and sulfur dioxide can damage plant leaves, interfere with photosynthesis, and stress plants, making them more susceptible to diseases. While the correlation is weak, it aligns with the ecological understanding that polluted environments are generally detrimental to agricultural productivity.

2.  **Relationship between `AQI` and `Overall_Total_Area` (Correlation: 0.44)**:
    *   **What it suggests**: There is a moderate positive correlation between a state's overall average AQI and its overall total cultivated area. This indicates that states with, on average, worse air quality (higher AQI) tend to have larger total areas dedicated to crop cultivation.
    *   **Why it might exist**: This is a more complex and potentially counter-intuitive relationship. It doesn't necessarily imply that poor air quality *causes* states to cultivate more land. Instead, it might suggest that states with larger overall land areas dedicated to agriculture might also encompass larger industrial and urban centers, which are primary contributors to higher average AQI. In essence, regions with extensive agricultural operations might often coincide with areas of significant economic activity and population density that also lead to higher pollution. The average AQI for a large, diverse state is influenced by all its components, potentially showing this positive correlation even if pollution negatively impacts local crop yields.
